```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mousedrug = pd.read_csv(mouse_drug_data_to_load)
clinicaltrial = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combineddf = pd.merge(mousedrug, clinicaltrial, on="Mouse ID", how='right')

# Display the data table for preview
combineddf.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
groupeddf = combineddf.groupby(['Drug','Timepoint'])["Tumor Volume (mm3)"]
tumvolmean = groupeddf.mean()

# Convert to DataFrame

tumvolmean = tumvolmean.reset_index()


# Preview DataFrame
tumvolmean
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumvolstderr = groupeddf.sem()

# Convert to DataFrame
tumvolstderrdf = tumvolstderr.reset_index()


# Preview DataFrame

tumvolstderrdf.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumvolmean_pivot = tumvolmean.pivot(index='Timepoint',columns = 'Drug', values = 'Tumor Volume (mm3)')
# Preview that Reformatting worked
tumvolmean_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
druglist = ['Capomulin','Infubinol','Ketapril','Placebo']
colorlist = ['red','blue','green','black']
markerlist = ['o','^','s','d']


for x in range(len(druglist)):
    semerror = tumvolstderrdf.loc[tumvolstderrdf["Drug"] == druglist[x], "Tumor Volume (mm3)"]
    plt.errorbar(tumvolmean_pivot.index,tumvolmean_pivot[druglist[x]],yerr=semerror,color=colorlist[x],marker=markerlist[x],label=druglist[x])

plt.legend(loc='best')
plt.grid(axis='y')
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time (Days)")
plt.title("Tumor Response to Treatment")

# Save the Figure
plt.savefig("tumor_responses.png")
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metdf = combineddf.groupby(["Drug", "Timepoint"])["Metastatic Sites"]
metmean = metdf.mean()

# Convert to DataFrame

metmean = metmean.reset_index()


# Preview DataFrame
metmean.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metmeanstderr = metdf.sem()

# Convert to DataFrame
metmeanstderrdf = metmeanstderr.reset_index()


# Preview DataFrame

metmeanstderrdf.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metmean_pivot = metmean.pivot(index='Timepoint',columns = 'Drug', values = 'Metastatic Sites')
# Preview that Reformatting worked
metmean_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
druglist = ['Capomulin','Infubinol','Ketapril','Placebo']
colorlist = ['red','blue','green','black']
markerlist = ['o','^','s','d']


for x in range(len(druglist)):
    semerror = metmeanstderrdf.loc[metmeanstderrdf["Drug"] == druglist[x], "Metastatic Sites"]
    plt.errorbar(metmean_pivot.index,metmean_pivot[druglist[x]],yerr=semerror,color=colorlist[x],marker=markerlist[x],label=druglist[x])

plt.legend(loc='best')
plt.grid(axis='y')
plt.ylabel("Met. Sites")
plt.xlabel("Treatment Duration (Days)")
plt.title("Metastatic Spread During Treatment")

# Save the Figure
plt.savefig("metastatic_spread.png")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survivors = combineddf.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"]
survivorcount = survivors.count()

# Convert to DataFrame
survivors = pd.DataFrame({"Mouse Count": survivorcount})
survivors = survivors.reset_index()

# Preview DataFrame
survivors.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survivors_pivot = survivors.pivot(index="Timepoint", columns="Drug")["Mouse Count"]

# Preview the Data Frame
survivors_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
druglist = ['Capomulin','Infubinol','Ketapril','Placebo']
colorlist = ['red','blue','green','black']
markerlist = ['o','^','s','d']


for x in range(len(druglist)):
    plt.plot(survivors_pivot[druglist[x]] / max(survivors_pivot[druglist[x]]) * 100,color=colorlist[x],marker=markerlist[x],label=druglist[x])

plt.legend(loc='best')
plt.grid()
plt.ylabel("Survival Rate (%)")
plt.xlabel("Time (Days)")
plt.title("Survival During Treatment")

# Save the Figure
plt.savefig("survival_rates.png")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumorpctchange =  100 * (tumvolmean_pivot.iloc[-1] - tumvolmean_pivot.iloc[0]) / tumvolmean_pivot.iloc[0]

# Display the data to confirm
tumorpctchange
```

```python
# Store all Relevant Percent Changes into a Tuple
tumorchanges = (tumorpctchange['Capomulin'],tumorpctchange['Infubinol'],tumorpctchange['Ketapril'],tumorpctchange['Placebo'])
# type(tumorchanges)


# Splice the data between passing and failing drugs
fig, ax = plt.subplots()
width = 1
tumorpass = ax.bar(np.arange(len(tumorchanges))[0], tumorchanges[0], width=-1, color='green',align="edge")
tumorfail = ax.bar(np.arange(len(tumorchanges))[1:], tumorchanges[1:], width=-1, color='red',align='edge')

# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')
plt.setp(ax, xticks=[0, 1, 2, 3], xticklabels=["Capomulin", "Infubinol", "Ketapril", "Placebo"],
        yticks=[-20, 0, 20, 40, 60])
ax.grid()

# Use functions to label the percentages of changes
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2, .1*height, "%d" %int(height)+ "%",
                ha='center', va='top', color="white")

# Call functions to implement the function calls
autolabel(tumorpass)
autolabel(tumorfail)

# Save the Figure
plt.savefig("passfail.png")
```

## Observations
#### 1. Capomulin is the most effective medication for tumors in mice. It has the best survival rate, it shrinks tumors, and has the least amount of Metastatic spread.
#### 2. While Infubinol helps with Metastatic spread, it kills the mice faster than any other. It kills mice faster than no medication at all (see Placebo) and does the opposite of what you want meciation to do.
#### 3. None of the drugs are effective at shrinking tumors or keeping mice alive other than Capomulin.
