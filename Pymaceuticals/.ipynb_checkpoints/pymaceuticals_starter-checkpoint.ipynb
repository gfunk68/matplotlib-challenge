```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mousedrug = pd.read_csv(mouse_drug_data_to_load)
clinicaltrial = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combineddf = pd.merge(mousedrug, clinicaltrial, on="Mouse ID", how='right')

# Display the data table for preview
combineddf.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
groupeddf = combineddf.groupby(['Drug','Timepoint'])["Tumor Volume (mm3)"]
tumvolmean = groupeddf.mean()

# Convert to DataFrame

tumvolmean = tumvolmean.reset_index()

# Preview DataFrame
tumvolmean
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumvolstderr = groupeddf.sem()

# Convert to DataFrame
tumvolstderrdf = tumvolstderr.reset_index()
# Preview DataFrame

tumvolstderrdf.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumvolmean_pivot = tumvolmean.pivot(index='Timepoint',columns = 'Drug', values = 'Tumor Volume (mm3)')
# Preview that Reformatting worked
tumvolmean_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
druglist = ['Capomulin','Infubinol','Ketapril','Placebo']
colorlist = ['red','blue','green','black']
markerlist = ['o','^','s','d']


for x in range(len(druglist)):
    plt.errorbar(tumvolmean_pivot.index,tumvolmean_pivot[druglist[x]],color=colorlist[x],marker=markerlist[x],label=druglist[x])

plt.legend(loc='best')
plt.grid(axis='y')


# Save the Figure

```

```python
# Show the Figure
plt.show()
```
